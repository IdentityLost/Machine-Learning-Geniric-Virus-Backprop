# Разделение на основные группы, такие как Название компетенции и т.д

```python
for row in range(len(data)):
  for z, t, r in zip(range(2, 28, 3), range(3, 28, 3), range(4, 28, 3)):
    if pd.isnull(data.iloc[row][z]) == True:
      break
    else:
      df.loc[len(df)] = np.nan
      df = df.shift()
      df.at[0, "название_компетенции_wsss"] = data.iloc[row][0]
      df.at[0, "описание_компетенции_wsss"] = data.iloc[row][1]
      df.at[0, "название_раздела_wsss"] = data.iloc[row][z]
      df.at[0, "значимость_раздела_wsss"] = data.iloc[row][t]
      df.at[0, "признаки_раздела_wsss"] = data.iloc[row][r]
  
df.drop(df.iloc[:,5:], inplace=True, axis = 1)
```

# Работа с текстом, удаление стоп слов

```python
import nltk
from nltk.corpus import stopwords

def clean_data(df):
  df = df.str.lower()
  df.replace(r'[^а-яА-Яa-zA-Z0-9]', ' ', regex = True, inplace = True)
  stop_words = stopwords.words("russian")
  df = df.apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
  return df
```

```python
nltk.download('stopwords')
```

```python
df["описание_компетенции_wsss"] = clean_data(df["описание_компетенции_wsss"])
df["название_раздела_wsss"] = clean_data(df["название_раздела_wsss"])
df["признаки_раздела_wsss"] = clean_data(df["признаки_раздела_wsss"])
df["название_компетенции_wsss"] = clean_data(df["название_компетенции_wsss"])
```

```python
df.head()
```

```python
df["значимость_раздела_wsss"] = pd.to_numeric(df["значимость_раздела_wsss"], errors = "coerce")
```

```python
df["значимость_раздела_wsss"] = df["значимость_раздела_wsss"].fillna(float(int(df["значимость_раздела_wsss"].median())))
```

# Создание выделителей

```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.feature_extraction.text import TfidfVectorizer

class TextSelector(BaseEstimator, TransformerMixin):
  def __init__(self, key):
    self.key = key
  def fit(self, X):
    return self
  def transform(self, X, y = None):
    return X[self.key]


class NumericalSelector(BaseEstimator, TransformerMixin):
  def __init__(self, key):
    self.key = key
  def fit(self, X):
    return self
  def transform(self, X, y=None):
    return X[[self.key]]
```

```python
from nltk.stem.snowball import SnowballStemmer
import re

stemmer = SnowballStemmer('russian')

def stemming_tokenizer(text):
  text = re.split('\W+', text)
  text = [stemmer.stem(word) for word in text]
  return text
```

```python
stop_words_s=stopwords.words('russian')

title = Pipeline([('selector', TextSelector(key='название_раздела_wsss')),
                  ('tfidf', TfidfVectorizer(stop_words=stop_words_s, tokenizer=stemming_tokenizer))
                ])

title.fit_transform(df)
```

```python
dictionary = Pipeline([('selector', TextSelector(key='описание_компетенции_wsss')),
                  ('tfidf', TfidfVectorizer(stop_words=stop_words_s, tokenizer=stemming_tokenizer))
                ])

dictionary.fit_transform(df)
```

```python
ves_title = Pipeline([('selector', NumericalSelector(key="значимость_раздела_wsss")),
                 ('tfidf', StandardScaler())
                ])
ves_title.fit_transform(df)
```

```python
titlefeat = Pipeline([('selector', TextSelector(key='признаки_раздела_wsss')),
                  ('tfidf', TfidfVectorizer(stop_words=stop_words_s, tokenizer=stemming_tokenizer))])
titlefeat.fit_transform(df)
```

# Начало кластеризации с помощью KMeans.

```python
from sklearn.pipeline import FeatureUnion
feat_full = FeatureUnion([('dictionary', dictionary),
                          ('titlefeat', titlefeat),
                          ('ves_title', ves_title)])

feat_process = Pipeline([('feat_full', feat_full)])
x = feat_process.fit_transform(df).toarray()
```

```python
from sklearn.metrics import silhouette_score
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2, random_state=0)
pca_feat = pca.fit_transform(x)
```

```python
from sklearn.cluster import KMeans
model_kmean = KMeans(n_clusters=7, n_jobs=-1, random_state=0)
model_kmean.fit(x)
pred = model_kmean.predict(x)
silhouette_score(pca_feat, pred)
```
