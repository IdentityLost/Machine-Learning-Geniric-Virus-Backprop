```python
from sklearn.metrics import classification_report
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import train_test_split
x_train, x_val, y_train, y_val = train_test_split(df2.iloc[:, 30:50], df2['SalePrice'], shuffle = True)
```

```python
x_val
```

```python
estimators = [100, 300, 500]
train_score = []
val_score = []

for estim in estimators:
  gbc = GradientBoostingClassifier(n_estimators=estim, verbose=3)
  gbc.fit(x_train, y_train)
  train_score.append(gbc.score(x_train, y_train))
  val_score.append(gbc.score(x_val, y_val))
```
